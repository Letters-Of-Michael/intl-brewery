```python
#Question 2
Truck_Size_volume_groups = brewery_df.groupby('Truck_Size')
TTruck_Size = Truck_Size_volume_groups[['Volume']].count()
TTruck_Size
```

```python
plt.bar(TTruck_Size.index, TTruck_Size.Volume);

plt.xlabel('Truck Size')
plt.ylabel('Volume Sold')

plt.title('Total Volume Sold By Truck');
```

```python
#Question 3
Month_volume_groups = brewery_df.groupby('Month')
Sum_months_volume_df = Month_volume_groups[['Volume']].sum()
Sum_months_volume_df
```

```python
plt.plot(Sum_months_volume_df.index, Sum_months_volume_df.Volume, marker = 'o');

plt.xlabel('Month')
plt.ylabel('Volume Sold')

plt.title('Total Volume Sold Per Month');
```

```python
#Question 4
Returned_groups = brewery_df.groupby('IsReturned')
Returned_df = Returned_groups[['Volume']].count()
Returned_df['%% volume'] = (Returned_df.Volume / 36282) * 100
Returned_df
```

```python
mylabels = ["N", "Y"]

plt.pie(Returned_df.Volume, labels = mylabels);
plt.legend()
plt.title('Percentage Volume Returned');
```

```python
#Question 5
Truck_Cat_groups = brewery_df.groupby('Truck_CAT')
Truck_Cat_df = Truck_Cat_groups[['Volume']].count()
Truck_Cat_df
```

```python
plt.bar(Truck_Cat_df.index, Truck_Cat_df.Volume);

plt.xlabel('Truck CAT')
plt.ylabel('Volume Sold')

plt.title('Total Volume Sold By Truck Category');
```

```python
#Question 6
Type_load_groups = brewery_df.groupby('Typeload')
Load_Type_df = Type_load_groups[['Volume']].count()
Load_Type_df
```

```python
plt.figure(figsize=(6, 3))
plt.bar(Load_Type_df.index, Load_Type_df.Volume);

plt.xlabel('Load_Type')
plt.ylabel('Volume Sold')

plt.title('Total Volume Sold By Load_Type');
```

```python
#Question 7
Month_volume_groups = brewery_df.groupby('Site')
Depot_volume_df = Month_volume_groups[['Volume']].sum()
Depot_volume_df
```

```python
plt.figure(figsize=(12, 6))
plt.bar(Depot_volume_df.index, Depot_volume_df.Volume);

plt.xlabel('Depot')
plt.ylabel('Volume Sold')

plt.title('Total Volume Sold By Depot');
```
